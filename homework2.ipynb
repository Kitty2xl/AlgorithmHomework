## Importing Libraries


```python
from collections import deque
from time import perf_counter
import matplotlib.pyplot as plt
import numpy as tf
```

## Fibonacci Top Down Algorithm (Memory Optimized)


```python
def TopDownFibonacci(n):

    counter = [0] * n

    buffer = [None] * n
    buffer[0] = 0
    buffer[1] = 1

    for i in range(n - 2):
        i += 2 

        if buffer[i] == None:
            buffer[i] = buffer[i - 2] + buffer[i - 1]
            counter[i] += 1

    return buffer[-1], counter
```

## Fibonacci Bottom Up Algorithm (Memory Optimized)


```python
def BottomUpFibonacci(n):

    f0 = 0
    f1 = 1

    buffer = deque(maxlen = 2)
    buffer.extend((f0, f1))

    if n not in [0, 1]:
        for _ in range(n - 1):

            total = sum(buffer)
            buffer.append(total)
        
    return buffer[-1]
```

## Question 1 : Comparing Algorithms

```python
N = 5000
axis = tf.arange(N)

TopDownTime = []  
BottomUpTime = []

for n in range(N):

    n += 2

    s = perf_counter()
    TopDownFibonacci(n)
    e = perf_counter()

    TopDownTime.append(e - s)

    s = perf_counter()
    BottomUpFibonacci(n)
    e = perf_counter()

    BottomUpTime.append(e - s)

plt.figure(figsize = (16, 8))
plt.xlabel('Fibonacci(N)')
plt.ylabel('Time (Seconds)')
plt.plot(axis, TopDownTime, label = 'Top Down')
plt.plot(axis, BottomUpTime, label = 'Bottom Up')
plt.legend()
plt.show()
```

## Question 2 : Calculating Overlap

```python
N = 5000
axis = tf.arange(N)
_, counter = TopDownFibonacci(N)

plt.figure(figsize = (16, 8))
plt.xlabel('Overlapped N')
plt.ylabel('Count')
plt.scatter(axis, counter, s = 0.25)
plt.legend()
plt.show()
```
